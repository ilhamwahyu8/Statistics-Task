```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# Get Year 2017

```python
dataset = pd.read_csv('Seasons_Stats.csv', index_col = 0)
dataset = dataset[dataset['Year'] == 2017.0]
# Drop Duplicated
dataset = dataset.drop_duplicates(subset=['Player'])
len(dataset)
```

```python
dataset.isna().sum()
```

```python
# Kolom blanl dan blank2 sengaja dihilangkan dikarenakan semua kolomnya berisi null
dataset = dataset.drop(['blanl', 'blank2'], axis=1)
# dataset = dataset.dropna()
```

```python
# Fungsi Mencari Player
def getDFYoungestPlayer(team):
    tempList = []
#     Mencari player pada tiap tim
    for item in team:
        tempDF = dataset.loc[dataset.Tm == item]
        ageMin = tempDF.Age.min()
#         Mencari player dengan umur termuda di team tersebut
        dfMinAgePlayer = tempDF.loc[tempDF.Age == ageMin][['Player', 'Age', 'Tm']]
        for i in range (len(dfMinAgePlayer)):
#             Menyatukan seluruh player tiap team
            tempList.append(dfMinAgePlayer.iloc[i])
    return pd.DataFrame(tempList)
def getDFOldestPlayer(team):
    tempList = []
#     Mencari player pada tiap tim
    for item in team:
        tempDF = dataset.loc[dataset.Tm == item]
        ageMax = tempDF.Age.max()
#         Mencari player dengan umur termuda di team tersebut
        dfMaxAgePlayer = tempDF.loc[tempDF.Age == ageMax][['Player', 'Age', 'Tm']]
        for i in range (len(dfMaxAgePlayer)):
#             Menyatukan seluruh player tiap team
            tempList.append(dfMaxAgePlayer.iloc[i])
    return pd.DataFrame(tempList)
```

# Youngest Age per Team

```python
# dataset.loc[dataset['Age'].groupby(dataset["Tm"]).idxmin()]
```

```python
teamList = dataset.Tm.unique()
dfYoungest = getDFYoungestPlayer(teamList)
dfYoungest.sort_values(by='Tm')
```

# Oldest Age per Team

```python
# dataset.loc[dataset['Age'].groupby(dataset["Tm"]).idxmax()]
```

```python
dfOldest = getDFOldestPlayer(teamList)
dfOldest.sort_values(by='Tm')
```

# Soal 4

```python
def createDFeachPos(posName, selectedAtr):
    tempDF = dataset[dataset['Pos'] == posName]
    tempDF = tempDF[selectedAtr]
    tempDF = tempDF.set_index('Player')
    tempDF.iloc[:, 2:] = MinMaxScaler().fit_transform(tempDF.iloc[:, 2:])
    tempKalimat = 'Overall ' + posName
#     -2 dikarenakan menghilangkan Pos dan Tm
    tempDF[tempKalimat] = tempDF.iloc[:, 2:].sum(axis=1)/ (len(tempDF.columns)-2)
    return tempDF
def createDFAllPos(selectedAtr):
    tempDF = dataset[selectedAtr]
    tempDF = tempDF.set_index('Player')
    tempDF.iloc[:, 2:] = MinMaxScaler().fit_transform(tempDF.iloc[:, 2:])
#     -2 dikarenakan menghilangkan Pos dan Tm
    tempDF['Overall'] = tempDF.iloc[:, 2:].sum(axis=1)/ (len(tempDF.columns)-2)
    return tempDF
```

```python
dfPG = createDFeachPos('PG', ['Player', 'Pos', 'Tm', 'AST', 'PTS'])
dfPG.sort_values(by=['Overall PG'], ascending=False).head(1)
```

```python
dfC = createDFeachPos('C', ['Player', 'Pos', 'Tm', 'TRB', 'BLK'])
dfC.sort_values(by=['Overall C'], ascending=False).head(1)
```

```python
dfSF = createDFeachPos('SF', ['Player', 'Pos', 'Tm', 'AST', 'PTS'])
dfSF.sort_values(by=['Overall SF'], ascending=False).head(1)
```

```python
dfSG = createDFeachPos('SG', ['Player', 'Pos', 'Tm', '3P', 'PTS'])
dfSG.sort_values(by=['Overall SG'], ascending=False).head(1)
```

```python
dfPF = createDFeachPos('PF', ['Player', 'Pos', 'Tm', 'TRB', 'PTS', 'BLK'])
dfPF.sort_values(by=['Overall PF'], ascending=False).head(1)
```

```python
dfAllPos = createDFAllPos(['Player', 'Pos', 'Tm', 'PTS',  'AST','TRB', 'PTS', 'BLK'])
dfAllPos.sort_values(by=['Overall'], ascending=False).head(1)
```
